##### 2. Change variable names so that they do include spaces
e.g: Activity Type -> Activity_Type       

```python
column_name_mapping = {
    'Activity Type': 'Activity_Type',
    'Avg HR': 'Avg_HR',
    'Max HR': 'Max_HR',
    'Aerobic TE': 'Aerobic_TE',
    'Avg Speed': 'Avg_Speed',
    'Max Speed': 'Max_Speed',
    'Total Ascent': 'Total_Ascent',
    'Total Descent': 'Total_Descent',
    'Min Elevation': 'Min_Elevation',
    'Max Elevation': 'Max_Elevation'
}

winter_sports_data = winter_sports_data.rename(columns=column_name_mapping)
```

##### 4. Remove data of activities that have frequency less than 5

##### 6. Remove all rows based on variables where Avg HR  = 0

##### 7. Split 'Title' in two columns: Location and Details using first space in the data

```python
winter_sports_data[['Location', 'Details']] = winter_sports_data['Title'].str.split(' ', 1, expand=True)
```

```python
winter_sports_data = winter_sports_data.drop(columns=['Title'])
```

```python
print(winter_sports_data.head())
```

##### 8. Make figure that shows the Location where is the highest frequency of activities

```python
location_counts = winter_sports_data['Location'].value_counts()
```

```python
most_frequent_location = location_counts.idxmax()
```

```python
plt.figure(figsize=(10, 6))
location_counts.plot(kind='bar', color='skyblue')
plt.title('Frequency of Activities by Location')
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
print("Location with the highest frequency:", most_frequent_location)
```

##### 9. Make scatter diagrams that show relation of variables 'Distance', 'Avg_HR', 'Calories' and 'Time'
- Hint: Use Seaborn pairplot
- Which variables seem to have highest correlation with each other?
- Which variables seem to have standardizaed distribution?

```python
selected_columns = ['Distance', 'Avg_HR', 'Calories', 'Time']
selected_data = winter_sports_data[selected_columns]
```

```python
selected_data['Time'] = pd.to_timedelta(selected_data['Time']).dt.total_seconds() / 60
```

```python
sns.pairplot(selected_data)
plt.suptitle('Scatter Diagrams for Distance, Avg_HR, Calories, and Time', y=1.02)
plt.show()
```

```python
correlation_matrix = selected_data.corr()
print(correlation_matrix)
```
